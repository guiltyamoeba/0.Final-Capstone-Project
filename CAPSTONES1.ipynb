## Merging the data

```python
merged_1 = pd.merge(appearances_df, games_df, on='game_id', how='inner')
```

```python
merged_2 = pd.merge(merged_1, games_events_df, on='game_id', how='inner')
```

```python
merged_3 = pd.merge(merged_2,games_lineups_df,left_on='player_id_x',right_on='player_id',how='inner')
```

```python
merged_4 = pd.merge(merged_3,players_df,on="player_id",how="inner")
merged_4.info()
```

```python
#Checking the number of missing values
merged_4.isnull().sum()
```

```python
merged_4.shape #knowing the no of rows and columns
```

#### Drop the unnecessary columns

```python
#Drop columns with more than 1000 missing values
columns_to_drop = merged_4.columns[merged_4.isnull().sum() > 1000]
treated_df = merged_4.drop(columns=columns_to_drop)

#drop unneccesary columns
columns_to_drop =["competition_id_y","game_id_y","player_id_y","date_x","date_y","player_name_x"]
treated_df=treated_df.drop(columns_to_drop,axis=1)
```

#### Missing value treatment

```python
#Treat remaining missing values
for column in treated_df.columns:
    if treated_df[column].isnull().sum() > 0:
        if  treated_df[column].dtype == 'object':  # Categorical
            treated_df[column].fillna(treated_df[column].mode()[0], inplace=True)  # fill with mode
        else:  # Numerical
            treated_df[column].fillna(treated_df[column].mean(), inplace=True) #fill with mean
```
